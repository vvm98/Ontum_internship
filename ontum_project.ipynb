### Extract date into day, hour, time

```python
# checking the data type of date columns
df.info()
```

```python
# Converting the date column to date time data type
df['sync_date'] = pd.to_datetime(df['sync_date'],format='%d-%m-%Y',errors='coerce')
df['date'] = pd.to_datetime(df['date'],format='%d-%m-%Y',errors='coerce')
```

```python
# Extract date into Year,month and day
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
column_index = (df.columns.get_loc('date'))+1
df.insert(column_index,'Year',df.pop('Year'))
df.insert(column_index+1,'Month',df.pop('Month'))
df.insert(column_index+2,'day',df.pop('day'))
```

```python
import pandas as pd

# Convert to Series first (if not already)
state_perc = pd.Series({
    "Karnataka": 57.792299,
    "Delhi": 18.081794,
    "Manipur": 5.887082,
    "Maharashtra": 5.788994,
    "Goa": 2.461650
})

# Calculate 'Other'
other_perc = 100 - state_perc.sum()

# Add "Other" to the Series
state_perc = pd.concat([state_perc, pd.Series({"Other": other_perc})])

print(state_perc)
```

```python
colors = sns.color_palette("pastel", len(state_perc))
plt.figure(figsize=(8, 8))
plt.pie(
    state_perc,
    labels=state_perc.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.savefig(r'D:\Desktop\internshipstates_total_usage_perc.png',facecolor=None,pad_inches=1.5,bbox_inches="tight")

plt.title("Percentage of Users by State (Top 5 + Other)")
plt.show()
```
