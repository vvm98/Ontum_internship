```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset of exam scores for students in different classes
data = {
    "class": ["A"]*20 + ["B"]*20 + ["C"]*20,
    "score": [55, 60, 62, 65, 66, 68, 70, 72, 74, 75,
              76, 77, 78, 80, 82, 85, 86, 88, 90, 95,   # Class A
              45, 48, 50, 52, 54, 55, 56, 57, 58, 60,
              62, 63, 64, 65, 67, 70, 72, 74, 76, 80,   # Class B
              30, 35, 40, 45, 50, 55, 60, 65, 70, 75,
              78, 80, 82, 84, 85, 87, 90, 92, 95, 100]  # Class C
}

df = pd.DataFrame(data)

df
```

```python
# Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x="class", y="score", data=df)
sns.stripplot(x="class", y="score", data=df, color="black",jitter=True,  alpha=0.6)

plt.title("Exam Scores Distribution by Class")
plt.show()
```
